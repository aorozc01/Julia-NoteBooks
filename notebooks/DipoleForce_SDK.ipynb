Here I explore the parameters required to create a SDK interferometer using the dipole force. The sequence goes like this

1. Prepare the atom in a ground state
2. apply a microwave $\pi/2$ pulse to prepare atom in a superposition
3. apply a pulse with a focused beam such that the atom experiences the maximum gradient
    * this will kick the atom in a direction that will depend on the state
4. wait a time T
5. apply a microwave $\pi$ pulse to swap atomic states
6. apply a pulse with a focused beam such that the atom experiences a field gradient
    * this will redirect the wavepackets by kicking the atom in an opposite direction as in line 3 above
7. wait a time T
8. apply a microwave $\pi/2$ pulse to create a superposition
9. apply a pulse with a focused beam such that the atom experiences a field gradient
    * this will be a kick in oppositie direction for each state

The dipole force on an atom is equal to the gradient of the dipole potential by the following equation $\vec{F} = -\vec{\Delta}U(r,z)$. The dipole potential can be written in terms of the atomic polarizability $\alpha$, $U = -\frac{1}{2\epsilon_0c}Re(\alpha) I(r)$.

The acceleration the atom experiences can be calculated using Newton's second law, $\vec{a}=\vec{F}/m$. 

### Define Constants

```julia
c = 3e8u"m/s"; # m/s
ϵ0 = 8.85e-12u"C/(V*m)"; # C/(V*m)
amu = 1.67e-27u"kg"; # kg
g = 10u"m/s^2";
kB = 1.38e-23u"J/K";
```

### Atom Characteristics

```julia
Γ = 2*π*4.575u"MHz"; # MHz 6P1/2 spontaneous emission rate
Isat = 2.5u"mW/cm^2"; # mW/cm^2; π-polarized light
m = 133*amu; # kg
λ = 895u"nm"; # D1 transition wavelength
ω0 = 2*π*c/λ;
ωhfs = 2*pi*9.193u"GHz";# Hz
```

### Beam Characteristics

```julia
ω₀ = 1u"μm"; # cm, beam radius/waist
s₀ = 10; # saturation parameter
I₀ = s₀*Isat; # W/cm^2
Δ = ωhfs/2;
```

```julia
function Polarizability(ω)
    C = 6*π*ϵ0*c^3;
    return C*(Γ/ω0^2)/(ω0^2-ω^2-im*(ω^3/ω0^2)*Γ)
end
function DipPotential(ω,I)
    A = -1/(2*ϵ0*c);
    return A*real(Polarizability(ω))*I;
end
function GaussianIntensity(r,ω₀,I₀)
    return I₀*exp.(-2*r.^2/ω₀^2);
end
function DipForce(r,ω,ω₀,I₀)
    U₀ = DipPotential(ω,I₀);
    B = -4*r/ω₀^2;
    return U₀*B.*exp.(-2*r.^2/ω₀^2)
end
```

```julia
r = [-3:0.1:3;].*u"μm";
F = DipForce(r,ω0-Δ,ω₀,I₀);
a = upreferred.(F)/m;
```

```julia
figure;
plot(ustrip(r),ustrip(a))
#vlines([-1,1],-3e12,3e12)
ylabel("Acceleration [m/s²]")
xlabel("Position [μm]");
plot(ustrip(r),3*ustrip(GaussianIntensity(r,ω₀,I₀)));
legend(["Acceleration","Beam Intensity"]);
```

From the graph above we see that the direction of the kick is dependent on which side of the beam center the atom is on. Also, in order to experience a kick the atom has to be within the beam profile as expected. The maximum kick occurs at $r_{max} = \pm\frac{\omega_0}{2}$. 

Knowing the maximum acceleration, how long should we wait to ensure we can drive atom back toward the center?

### Kinematic Eqautions

```julia
t = [0:0.01:200;].*u"μs";
amax = maximum(a);
x = upreferred.((1/2)*amax.*t.^2);
```

```julia
figure;
x = uconvert.(u"μm",x)
plot(ustrip(t),ustrip(x))
ylabel("Position [μm]")
xlabel("Time [μs]");
#hlines(ustrip(ω₀)/2,0,200)
```

The above plot assumes the atom starts off at the maximum location to experience the largest acceleration. It then proceeds to accelerate and change position. In this case it starts off at the position of $\omega_0/2$ and thus is constrained to move a maximum distance of $\omega_0/2$ before it leaves the beam intensity region.
