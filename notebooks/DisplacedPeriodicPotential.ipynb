```julia
using QuantumOptics
using PyPlot
using Unitful
```

```julia
xmin = -10;
xmax = 10;
xpoints = 100;
b_position = PositionBasis(xmin,xmax,xpoints) 
X = position(b_position);
P = momentum(b_position);
```

```julia
Vp = exp(im*dense(X))/2;
Vm = exp(-im*dense(X))/2;
V = (Vp.+Vm)^2;
```

```julia
H = P^2/2+V-X;
```

```julia
ψ0 = basisstate(b_position,51);
expect(X,ψ0)
```

```julia
T = [0:0.01:3;]
tout, Ψt = timeevolution.schroedinger(T, ψ0, H);
```

```julia
exp_pos = real(expect(P,Ψt));
figure;
plot(T,exp_pos)
```
